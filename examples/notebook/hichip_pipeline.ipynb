# HiChIP Pipeline

```python
import chr3d as c3d
```

## Set Input Files

```python
genome = '/path/to/hg38.fa'
fragments = '/path/to/mboi_fragments.bed'
r1 = '/path/to/sample_R1.fastq.gz'
r2 = '/path/to/sample_R2.fastq.gz'
output_dir = './hichip_results'
```

## Run Pipeline

```python
pipeline = c3d.HiChIPPipeline(
    genome_index=genome,
    fragment_bed=fragments
)
```

```python
pipeline.run(
    fastq_r1=r1,
    fastq_r2=r2,
    output_dir=output_dir
)
```
