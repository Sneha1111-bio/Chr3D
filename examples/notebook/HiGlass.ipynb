# Visualizing Loops (Arcs) along with other Tracks


```python
# For Arcs from typing import Literal
from typing import Literal

# ── ArcsTrack Plugin Definition ───────────────────────────────────────────────
class ArcsTrack(hg.PluginTrack):
    type: Literal['1d-arcs'] = '1d-arcs'
    plugin_url: str = 'https://unpkg.com/higlass-arcs/dist/higlass-arcs.min.js'
```

```python
# Adding Additional Track genes and chrom sizes
COOL_FILE = "/app/trash/higlass/contact_matrix.mcool"
CHROM_SIZE = "/app/trash/higlass/hg38.chrom.sizes"
LOOPS = "/app/trash/higlass/loops.bedpedb"

# read all the file in higlass object
contact_matrix_track = hg.cooler(COOL_FILE)
hg_chrom_size = hg.chromsizes(CHROM_SIZE)
hg_loops = hg.bed2ddb(LOOPS)

# Decide Track Types
linear_heatmap = contact_matrix_track.track("linear-heatmap")
chrom_sizes = hg_chrom_size.track("horizontal-chromosome-labels")
_arc_track = hg_loops.track("2d-rectangle-domains")


arcs = ArcsTrack(
    type='1d-arcs',
    tilesetUid=_arc_track.tilesetUid,
    server=_arc_track.server,
    height=150,
    plugin_url='https://unpkg.com/higlass-arcs/dist/higlass-arcs.min.js',
    options={
        'strokeColor': 'red',
        'strokeWidth': 2.0,
        'strokeOpacity': 0.8,
        'arcStyle': 'circle',
        'flip1D': 'yes',
        'name': 'ChromSpec Interactions'
    }
)

view = hg.view(
    (chrom_sizes,"top"),
    (linear_heatmap,"center"),
    (arcs,"bottom")
    )

view
```
