# Single-Nucleus Hi-C Pipeline

```python
import chr3d as c3d
```

## Set Input Files

```python
genome = '/path/to/hg38.fa'
chrom_sizes = '/path/to/hg38.chrom.sizes'
manifest = '/path/to/cells.tsv'
output_dir = './snhic_results'
```

## Run Pipeline

```python
pipeline = c3d.SnHiCPipeline(
    genome_index=genome,
    chrom_sizes=chrom_sizes
)
```

```python
# Read manifest file
import csv
cells = []
with open(manifest) as f:
    for row in csv.reader(f, delimiter='\t'):
        cells.append((row[0], row[1], row[2]))
```

```python
pipeline.run(
    cells=cells,
    output_dir=output_dir
)
```
