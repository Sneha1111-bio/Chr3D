# ChIA-PET Pipeline

```python
import chr3d as c3d
```

## Set Input Files

```python
genome = '/path/to/hg38.fa'
r1 = '/path/to/sample_R1.fastq.gz'
r2 = '/path/to/sample_R2.fastq.gz'
linkers = ['ACGCGATATCGCG']
output_dir = './chiapet_results'
```

## Run Pipeline

```python
pipeline = c3d.ChiaPetPipeline(
    genome_index=genome,
    linkers=linkers
)
```

```python
pipeline.run(
    fastq_r1=r1,
    fastq_r2=r2,
    output_dir=output_dir
)
```
